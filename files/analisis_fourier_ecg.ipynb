```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import json
```

```python
with open("ecg.txt","r") as infile:
    datos=infile.readlines()
```

```python
datos
```

```python
arch3=[]

for linea in datos:
    arch2=linea.strip()
    if(len(arch2)!=0):
        x=arch2
        arch3.append((int(x)*5)/1023)
```

```python
arch3
```

```python
with open('ydataecg.json', 'w') as outfile:
    json.dump(arch3, outfile)
```

```python
with open('ydataecg.json', 'r') as file:
        ydata = json.load(file)
```

```python
ydata
```

```python
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 30,
        }
```

```python
fig = plt.figure(figsize=(30,10))
ax = fig.add_subplot(111)
#ax.scatter(ejex, poblacion, lw = 2 )
#ax.plot(ydata,color='k', alpha = 1, linewidth= 1.0 )
#ax.scatter(ejex, ejem, lw = 2 )
plt.xlabel('tiempo en 1/100 seg', fontdict=font)
plt.ylabel('volt', fontdict=font)
x=sp.linspace(0,30,len(ydata))
plt.xlim([0,30])
y=ydata
plt.ylim([0,5])
plt.plot(x,y,color='k')
# Tweak spacing to prevent clipping of ylabel
plt.subplots_adjust(left=0.15)
plt.title("Ritmo", fontdict=font)
plt.savefig('hato.pdf')
plt.show()
```

```python
N = len(ydata)
Y = np.fft.fft(ydata)
freq = np.fft.fftfreq(N)
ind_freq =np.arange(1, N//2 +1)
N
```

```python
freq
```

```python
freq[-ind_freq]
```

```python
psd = np.abs(Y[ind_freq])**2 + np.abs(Y[-ind_freq])**2
```

```python
ind = np.where(psd > 1000)
mainf=freq[ind_freq[ind]]
len(mainf)
```

```python
fig = plt.figure(figsize=(30,10))
ax = fig.add_subplot(111)
for i in range(len(mainf)):
    ax.axvline(mainf[i],color='red',alpha = 0.75 )
ax.set_xlim([0, 0.5])
ax.scatter(freq[ind_freq],psd,color='black', s = 90.25, alpha = 0.55)
plt.show()
```
