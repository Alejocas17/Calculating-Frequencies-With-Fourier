```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import json
```

```python
with open("pulso.txt","r") as infile:
    datos=infile.readlines()
```

Leemos el archivo output.txt, y leemos cada linea del archivo. 

```python
datos
```

```python

lectura=[]

for linea in datos:
    dato=linea.strip()
    if(len(dato)!=0):
        x=dato
        lectura.append((int(x)*5)/1023)
    
```

```python
lectura

```

```python
with open('ydata.json', 'w') as outfile:
    json.dump(lectura, outfile)
```

```python
with open('ydata.json', 'r') as file:
        ydata = json.load(file)
```

```python
ydata
```

```python
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 30,
        }
```

```python
fig = plt.figure(figsize=(30,10))
ax = fig.add_subplot(111)
#ax.scatter(ejex, poblacion, lw = 2 )
#ax.plot(ydata,color='k', alpha = 1, linewidth= 1.0 )
#ax.scatter(ejex, ejem, lw = 2 )
plt.xlabel('tiempo en 1/100 seg', fontdict=font)
plt.ylabel('volt', fontdict=font)
x=sp.linspace(0,60,len(ydata))
plt.xlim([0,60])
y=ydata
plt.ylim([0,5])
plt.plot(x,y,color='k')
# Tweak spacing to prevent clipping of ylabel
plt.subplots_adjust(left=0.15)
plt.title("Pulsaciones", fontdict=font)
plt.savefig('hato.pdf')
plt.show()
```

```python
N = len(ydata)
Y = np.fft.fft(ydata)
freq = np.fft.fftfreq(N)
ind_freq =np.arange(1, N//2 +1)
```

```python
freq
```

```python
freq[-ind_freq]
```

```python
psd = np.abs(Y[ind_freq])**2 + np.abs(Y[-ind_freq])**2
```

```python
ind = np.where(psd > 450)
mainf=freq[ind_freq[ind]]
len(mainf)
```

```python
psd
```

```python
fig = plt.figure(figsize=(30,10))
ax = fig.add_subplot(111)
for i in range(len(mainf)):
    ax.axvline(mainf[i],color='red',alpha = 0.75 )
ax.set_xlim([0, 0.1])
ax.scatter(freq[ind_freq],psd,color='black', s = 90.25, alpha = 0.55)
plt.show()
```
